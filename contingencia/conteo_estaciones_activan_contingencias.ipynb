```python
import pandas as pd
from collections import defaultdict, Counter
```

```python
dir = "C:/Users/valer/Documents/CIC/doctorado/A24/Tesis/contingencias.csv"
df = pd.read_csv(dir, encoding = "Latin1")
df
```

```python
zonas_cont = []
estaciones_cont = []
for i in range(len(df)):
    zonas = df.iloc[i].zona
    zonas_lista = zonas.split('/')
    zonas_lista = [palabra.strip().upper() for palabra in zonas_lista]
    estaciones = df.iloc[i].estacion_inicio
    estaciones_lista = estaciones.split('/')
    estaciones_lista = [palabra.strip().upper() for palabra in estaciones_lista]
    if len(zonas_lista) == len(estaciones_lista):
        for j in range(len(zonas_lista)):
            zonas_cont.append(zonas_lista[j])
            estaciones_cont.append(estaciones_lista[j])
```

```python
estaciones_cont = [s.replace('-', '').replace(' ', '') for s in estaciones_cont]
estaciones_cont = [s.replace('ENEPACATLÁN', 'FESACATLÁN') for s in estaciones_cont]
```

```python
# Paso 1: Crear un diccionario para almacenar valores por categoría
valores_por_categoria = defaultdict(list)

# Paso 2: Recorrer ambas listas y agregar valores al diccionario según la categoría
for categoria, valor in zip(zonas_cont, estaciones_cont):
    valores_por_categoria[categoria].append(valor)

# Paso 3: Contar las ocurrencias de cada valor por categoría y encontrar el más común
resultados = {}

for categoria, lista_valores in valores_por_categoria.items():
    conteo = Counter(lista_valores)
    valor_mas_comun, cantidad = conteo.most_common(1)[0]
    resultados[categoria] = (valor_mas_comun, cantidad)

# Mostrar resultados
for categoria, (valor, cantidad) in resultados.items():
    print(f"Categoría {categoria}: Valor más común es {valor} con {cantidad} repeticiones.")
```

```python
# Contar las ocurrencias de cada valor en la lista
conteo = Counter(estaciones_cont)

# Obtener todos los valores y sus conteos, ordenados de mayor a menor frecuencia
valores_ordenados = conteo.most_common()

# Mostrar resultados
valor_mas_comun, cantidad = conteo.most_common(1)[0]
print(f"El valor más común es {valor_mas_comun} con {cantidad} repeticiones.")

# Mostrar resultados
print("Valores ordenados por frecuencia (de mayor a menor):")
for valor, cantidad in valores_ordenados:
    #print(f"Valor {valor}: {cantidad} repeticiones")
    print(valor, cantidad)
```
