```python
def imagen_transparente(ruta_imagen):
    imagen = Image.open(ruta_imagen)
    # Obtener el canal alfa (transparencia)
    canal_alfa = imagen.split()[3]
    return all(valor == 0 for valor in canal_alfa.getdata())
```

```python
import geopandas as gpd
from osgeo import ogr
import fiona

# Ruta del archivo PBF
ruta_pbf = 'C:/Users/valer/Documents/CIC/doctorado/air_pollution/traffic_flow/raster_tiles/respuesta.pbf'
# Ruta de salida del archivo GeoJSON
ruta_geojson = 'salida.geojson'

# Leer el archivo PBF con geopandas
gdf = gpd.read_file(fiona.Collection(ruta_pbf, vsi='OSM'))

# Guardar el GeoDataFrame como GeoJSON
gdf.to_file(ruta_geojson, driver='GeoJSON')

print(f'Conversión completada. GeoJSON guardado en: {ruta_geojson}')
```

```python
from osgeo import ogr

# Ruta del archivo PBF
#ruta_pbf = 'ruta/a/tu/archivo.pbf'

# Ruta de salida del archivo GeoJSON
#ruta_geojson = 'ruta/a/tu/salida.geojson'

# Abrir el archivo PBF con GDAL (ogr)
datasource = ogr.Open(ruta_pbf)
layer = datasource.GetLayer()

# Crear un archivo GeoJSON y escribir la capa en él
driver = ogr.GetDriverByName('GeoJSON')
geojson_datasource = driver.CreateDataSource(ruta_geojson)
geojson_layer = geojson_datasource.CreateLayer('capa', srs=layer.GetSpatialRef(), geom_type=layer.GetGeomType())

# Copiar los campos de la capa original a la capa GeoJSON
for i in range(layer.GetLayerDefn().GetFieldCount()):
    field_defn = layer.GetLayerDefn().GetFieldDefn(i)
    geojson_layer.CreateField(field_defn)

# Copiar las características (features) de la capa original a la capa GeoJSON
for feature in layer:
    geojson_layer.CreateFeature(feature)

# Cerrar los datasources
datasource = None
geojson_datasource = None

print(f'Conversión completada. GeoJSON guardado en: {ruta_geojson}')
```
