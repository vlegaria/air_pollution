```python
import osmnx as ox

# Definir el nombre del lugar
place_name = "Ciudad de Mxico, Mexico"

# Descargar el grafo para conduccin (puedes cambiar 'drive' por otro tipo de red si lo deseas)
graph = ox.graph_from_place(place_name, network_type='drive')

# Mostrar el mapa para verificar que todo est funcionando
ox.plot_graph(ox.project_graph(graph))
```

```python
# Obtener los nodos y las aristas (edges) del grafo
nodes, edges = ox.graph_to_gdfs(graph)

# Mostrar los identificadores de las calles (aristas)
print(edges[['osmid', 'name']])  # 'osmid' es el identificador nico de cada calle
```

```python
edges.set_crs("EPSG:4326", inplace=True)  
edges.to_file('osm_cdmx.geojson', driver='GeoJSON')
```
